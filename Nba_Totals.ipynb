```python

import pandas as pd

spreadsheet2023 = pd.read_csv('/Users/lilgl/Downloads/BrainStation_Final_Project/nba2022_2023.csv')

# Assuming you have a DataFrame named 'spreadsheet2023'
db = spreadsheet2023


# Calculate PTS/G by dividing PTS by G
db["PTS"] = db["PTS"] / db["G"]
db = db.sort_values(by="PTS", ascending=False)

# Reset the index of the DataFrame
db.reset_index(drop=True, inplace=True)

# Get the index of each item and add one to it
new_column = db.index + 1

# Assign the new column to the DataFrame
db['Rk'] = new_column

# Generate the scores based on the ranking, assign 0 if negative
db['Score'] = [max(10 - rank, 0) for rank in range(len(db))]


db
# # Iterate over the rows of the DataFrame and print player information with PTS/G
# for index, row in db.iterrows():
#     print(f"Player: {row['Player']}, PTS/G: {row['PTS/G']}")
```

```python
# Filter the DataFrame to show players with score > 0
bt = db[db['Score'] > 0]

bt
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with columns: 'Player', 'PTS', 'AST', 'TRB'

# Sort the DataFrame by each category in descending order
df_sorted_pts = db.sort_values(by='PTS', ascending=False)
df_sorted_ast = db.sort_values(by='AST', ascending=False)
df_sorted_trb = db.sort_values(by='TRB', ascending=False)

# Select the top 10 players in each category
top_10_pts = df_sorted_pts.head(10)
top_10_ast = df_sorted_ast.head(10)
top_10_trb = df_sorted_trb.head(10)

# Create a combined DataFrame with the top 10 players in each category
top_10_combined = pd.concat([top_10_pts, top_10_ast, top_10_trb]).drop_duplicates()

# Sort the combined DataFrame by each category in descending order
top_10_combined_sorted_pts = top_10_combined.sort_values(by='PTS', ascending=False)
top_10_combined_sorted_ast = top_10_combined.sort_values(by='AST', ascending=False)
top_10_combined_sorted_trb = top_10_combined.sort_values(by='TRB', ascending=False)

# Plot the prediction graph
plt.figure(figsize=(12, 6))
plt.plot(top_10_combined_sorted_pts['Player'], top_10_combined_sorted_pts['PTS'], label='Points')
plt.plot(top_10_combined_sorted_ast['Player'], top_10_combined_sorted_ast['AST'], label='Assists')
plt.plot(top_10_combined_sorted_trb['Player'], top_10_combined_sorted_trb['TRB'], label='Rebounds')
plt.xlabel('Player')
plt.ylabel('Value')
plt.title('Top 10 Players in Points, Assists, and Rebounds')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
X = bt['PTS'].values.reshape(-1, 1)  # Annoying hack you have to do when you have one independent variable
y = bt['Score']
```

```python
bt_model = linear_model.LogisticRegression(solver='liblinear')
bt_model.fit(X, y)
```

```python
Xnew = np.arange(1, 30).reshape(-1, 1)
bt_model.predict(Xnew)
```

```python
beta_0 = bt_model.intercept_[0]
print(beta_0)

beta_1 = bt_model.coef_[0][0]
print(beta_1)
```

```python
def predict_probability(x):
    return 1/(1 + np.exp(-(beta_0 + beta_1*x)))

[predict_probability(x) for x in Xnew]
```
